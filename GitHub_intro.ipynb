```python
# Import necessary libraries  
#In Command Prompt, run "py -m pip install pyodbc" and then "py -m pip install pandas"
import pandas as pd  
import pyodbc  
from IPython.display import HTML, display  

# Define the connection string  
conn_str = (  
    "DRIVER={ODBC Driver 18 for SQL Server};"  
    "SERVER=(localdb)\\LocalServer1;"  
    "DATABASE=master;"  # Replace with your actual database name  
    "Trusted_Connection=yes;"  
)  

# Establish a connection to the database  
conn = pyodbc.connect(conn_str)  
  
# Write the SQL query to select the first 10 records  
query = "SELECT TOP 10 * FROM PropertySales;"  
  
# Use pandas to read the SQL query into a DataFrame  
df = pd.read_sql(query, conn)  
  
# Display the DataFrame  
#print(df)  

# Convert DataFrame to HTML  
html = df.to_html(classes='table-striped', escape=False)  
  
# Display the HTML with a scrollable container  
display(HTML(f'<div style="overflow-x: auto; width:100%">{html}</div>'))  
  
# Close the connection  
conn.close()  

```

```python
import pandas as pd  
import folium  
import geopandas as gpd  
  
# Load your data into a DataFrame  
data = {  
    'Region': ['EAST ANGLIA', 'EAST MIDLANDS'],  
    'Overall_Average': [333505, 266295],  
    'Country': ['UK', 'UK']  
}  
df = pd.DataFrame(data)  
  
# Load UK regions GeoJSON (example URL, replace with your actual GeoJSON file)  
# GeoJSON files can be found from various sources or created from shapefiles  
geo_json_url ='G:\My Drive\Old PC\AT\scrum session\lad.json'  
  
# Load GeoJSON data into a GeoDataFrame  
gdf = gpd.read_file(geo_json_url)  
  
# Merge your DataFrame with the GeoDataFrame  
merged = gdf.set_index('region').join(df.set_index('Region'))  
  
# Initialize a folium map centered around the UK  
m = folium.Map(location=[54.5, -3.0], zoom_start=5)  
  
# Add the choropleth layer  
folium.Choropleth(  
    geo_data=merged,  
    data=merged,  
    columns=['region', 'Overall_Average'],  
    key_on='feature.properties.region',  
    fill_color='YlGnBu',  
    fill_opacity=0.7,  
    line_opacity=0.2,  
    legend_name='Overall Average Prices'  
).add_to(m)  
  
# Display the map  
m.save('uk_map.html')  


```

### GitHub Repository Structure  
  
**Repository**

**(1-3: this session)**  
1. **Branches (Parallel development)**  
    - Main (Production ready state)  
    - Feature Branches (New features, branched off the "Main" branch)
    - Bugfix Branches (Bug fixing, branched off the "Main" branch)
    
2. **Commits (record of changes made to the repository/files, project history)**  

3. **Pull Requests (Review changes, before integrating to the "Main" branch)**  
    - Code Review (Reviewers can provide feedback, request changes, or approve the changes.)
    - Discussions (Team members can ask questions, make comments)
    - Merge (Once a pull request is reviewed and approved, it can be merged into the main branch)




**(4-12: FYI)**

4. **Issues**  
    - Bug Reports  
    - Feature Requests  
    - Tasks  
5. **README.md**  
6. **License**  
7. **gitignore**  
8. **Wikis**  
9. **Projects**  
    - Kanban Board  
    - Columns: To Do, In Progress, Done  
    - Cards  
10. **Actions (CI/CD Workflows)**  
    - Workflows  
    - Jobs  
    - Steps  
11. **Releases**  
    - Tags  
    - Release Notes  
12. **Security**  
    - Dependabot Alerts  
    - Security Policies
