```python
#In Command Prompt, run "py -m pip install pyodbc" and then "py -m pip install pandas"

# Import necessary libraries 
import pandas as pd  
import pyodbc  
import warnings  

from IPython.display import HTML, display  

# hide a warning
warnings.filterwarnings("ignore", message="pandas only supports SQLAlchemy connectable")  


# Define the connection string  
conn_str = (  
    "DRIVER={ODBC Driver 18 for SQL Server};"  
    "SERVER=(localdb)\\LocalServer1;"  #enter a server name
    "DATABASE=master;"  # enter a database name
    "Trusted_Connection=yes;"  
)  

# Establish a connection to the database  
conn = pyodbc.connect(conn_str)  
  
# select the first 10 records  
query = "SELECT TOP 10 * FROM PropertySales;"  
  
# Use pandas to read the SQL query into a DataFrame  
df = pd.read_sql(query, conn)  
  
# Display the DataFrame if preferred 
#print(df)  

# Convert DataFrame to HTML  
html = df.to_html(classes='table-striped', escape=False)  
  
# Display the HTML with a scrollable container  
display(HTML(f'<div style="overflow-x: auto; width:100%">{html}</div>'))  
  
# Close the connection  
conn.close()  

```

### GitHub Repository Structure  
  
**Repository**

**(1-3: this session)**  
1. **Branches (Parallel development)**  
    - Main (Production ready state)  
    - Feature Branches (New features, branched off the "Main" branch)
    - Bugfix Branches (Bug fixing, branched off the "Main" branch)
    
2. **Commits (record of changes made to the repository/files, project history)**  

3. **Pull Requests (Review changes, before integrating to the "Main" branch)**  
    - Code Review (Reviewers can provide feedback, request changes, or approve the changes.)
    - Discussions (Team members can ask questions, make comments)
    - Merge (Once a pull request is reviewed and approved, it can be merged into the main branch)
